```python
from urllib.request import urlopen # Comando de requisição para browser SOMENTE abrir o URL utilizando a biblioteca 
# padrão do Python
from bs4 import BeautifulSoup # O objeto mais comumente usado na biblioteca é importado
html = urlopen("http://www.pythonscraping.com/exercises/exercise1.html")# Comando para abrir o site chamado de html com o 
# endereço entre ()
bsObj = BeautifulSoup(html.read());# Usamos o método read para lermos o conteudo do HTML e guardamos o conteúdo em bsObj 
# que é transformado em objeto BeautifulSoup com a estrutura...celula abaixo
print(bsObj.h1)
print(bsObj.html.body.h1)
print(bsObj.body.h1)
print(bsObj.html.h1)
```
